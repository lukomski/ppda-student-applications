Remove from dataset applications which does not have status which results from dean's office decision

```python
class Decision:
    descriptions = (
        'Zamkniêty - decyzja pozytywna',
        'Zamkniêty - decyzja negatywna',
        'Zamkniêty - inne',
        'Zamkniêty - odrzucony formalnie',
        'Zwrócony do korekty',
        'Decyzja pozytywna - oczekuje na realizacjê'
    )
    
    @staticmethod
    def getClassId(description):
        return Decision.descriptions.index(description)
    
    @staticmethod
    def getDescription(classId: int):
        return Decision.descriptions[classId]
```

```python
sa = sa.loc[sa['Application status'].isin(Decision.descriptions)]
sa['Application status'].unique()
```

```python
decision_column = sa.apply(lambda row: Decision.getClassId(row['Application status']), axis=1)
sa['decision'] = decision_column
```
